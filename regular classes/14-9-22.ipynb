```python
import pandas as pd 
import numpy as np
df = pd.DataFrame(np.random.rand(20,5), columns=['Jan', 'Feb', 'Mar', 'April', 'May'])
df.plot(kind='hist',bins=20,figsize=(20,10), color=['r','b','g','y','c']),legend(bbox_to_anchor=(1.1, 1))
#bbox_to anchor is use to play labels outside the graph
#figsize is used to make the graph big and if we exclude it it will be small
#color is used to change color
#to make the graph stacked we gave stacked=True
#random is used to give random value random.rand(rows,columns)
#barh is used for horizontal bar graph
#hist -> histrogram and bins is used to change the width of the bar
#bar -> used to give bar graph
#pie ->used to give pie chart
```

```python
import pandas as pd
import numpy as np
salesMen = ['Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila']
Mobile_Sales = [2540, 1370, 1320, 2000, 2100, 2150]
TV_Sales = [2200, 1900, 2150, 1850, 1770, 2000]
df = pd.DataFrame()
df ['Name'] =salesMen
df ['Mobile_Sales'] = Mobile_Sales
df['TV_Sales']=TV_Sales
df.set_index("Name",drop=True, inplace=True)
df
```

```python
df.plot.bar( figsize=(20, 10), rot=0).legend(bbox_to_anchor=(1.1,1))
```

```python
df.plot.pie(subplots=True)
```

```python
df.plot.bar(stacked=True, figsize=(20, 10), rot=0).legend(bbox_to_anchor=(1.1, 1))
```

```python
df.plot.hist(bins=50)
```
